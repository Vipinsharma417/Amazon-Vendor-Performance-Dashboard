```python
import numpy as np
import pandas as pd
```

```python
import os
```

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine("sqlite:///inventory.db")
```

```python
for file in os.listdir('data'):
    if ".csv" in file:
        df = pd.read_csv('data/'+file)
        print(df.shape)
```

```python
def ingest_db(df, table_name, engine):
    df.to_sql(table_name, con= engine, if_exists='replace', index=False)
```

```python


def ingest_db_chunked(file_path, table_name, engine, chunksize=100_000):
    print(f"Starting ingestion: {table_name}")
    try:
        for i, chunk in enumerate(pd.read_csv(file_path, chunksize=chunksize)):
            chunk.to_sql(name=table_name, con=engine, if_exists='append', index=False)
            print(f"  --> Inserted chunk {i+1}")
        print(f"✅ Finished ingestion: {table_name}")
    except Exception as e:
        print(f"❌ Failed to ingest {table_name}: {e}")

processed_files = set()

for file in os.listdir('data'):
    if file.endswith(".csv") and file not in processed_files:
        processed_files.add(file)
        file_path = os.path.join('data', file)
        table_name = file[:-4]  # Remove ".csv"
        ingest_db_chunked(file_path, table_name, engine)
```
